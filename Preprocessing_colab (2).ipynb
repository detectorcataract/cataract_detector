```python
import tensorflow as tf
import os

# Paths
INPUT_DIR = "/content/drive/MyDrive/cataract_dataset"
OUTPUT_DIR = "/content/drive/MyDrive/preprocessed_data_pad"

# Image size
IMG_SIZE = (224, 224)

# Process train, valid and test folders
for split in ["train", "valid", "test"]:

    input_split_path = os.path.join(INPUT_DIR, split)

    for class_name in os.listdir(input_split_path):

        input_class_path = os.path.join(
            input_split_path,
            class_name
        )

        output_class_path = os.path.join(
            OUTPUT_DIR,
            split,
            class_name
        )

        # Create output folders automatically
        os.makedirs(output_class_path, exist_ok=True)

        image_files = os.listdir(input_class_path)

        print(f"Processing {split}/{class_name}...")

        for image_file in image_files:

            try:

                input_image_path = os.path.join(
                    input_class_path,
                    image_file
                )

                # Read image
                image = tf.io.read_file(
                    input_image_path
                )

                # Decode image as RGB
                image = tf.image.decode_image(
                    image,
                    channels=3
                )

                # Resize
                image = tf.image.resize(
                    image,
                    IMG_SIZE
                )

                # Brightness enhancement
                image = tf.image.adjust_brightness(
                    image,
                    delta=0.05
                )

                # Contrast enhancement
                image = tf.image.adjust_contrast(
                    image,
                    contrast_factor=1.1
                )

                # Keep values valid
                image = tf.clip_by_value(
                    image,
                    0,
                    255
                )

                # Convert back to uint8
                image = tf.cast(
                    image,
                    tf.uint8
                )

                # Save path
                output_image_path = os.path.join(
                    output_class_path,
                    image_file
                )

                # Save image
                tf.io.write_file(
                    output_image_path,
                    tf.io.encode_jpeg(image)
                )

            except Exception as e:

                print(
                    f"Error processing {image_file}: {e}"
                )

print("\nPreprocessing Completed Successfully!")
```
